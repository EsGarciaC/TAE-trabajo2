```python
import io
from google.colab import files
import pandas as pd

uploaded = files.upload()
loan_data1 = pd.read_csv(io.BytesIO(uploaded["loan_data_2007_2014-1.csv"]))
```

```python
uploaded = files.upload()
loan_data2 = pd.read_csv(io.BytesIO(uploaded["loan_data_2007_2014-2.csv"]))
```

```python
uploaded = files.upload()
loan_data3 = pd.read_csv(io.BytesIO(uploaded["loan_data_2007_2014-3.csv"]))
```

```python
loan_data_save = pd.concat([loan_data1, loan_data2, loan_data3],                  # Append two pandas DataFrames
                      ignore_index = True,
                      sort = False)
loan_data_save
```

```python
pd.options.display.max_columns = None
```

```python
loan_data = loan_data_save.copy()
loan_data
```

```python
loan_data.dropna(axis=1, how="all", inplace=True)
loan_data
```

```python
loan_data.columns
```

```python
loan_data.drop(labels='policy_code', axis=1, inplace=True)
loan_data.drop(labels='funded_amnt', axis=1, inplace=True)
loan_data.drop(labels='funded_amnt_inv', axis=1, inplace=True)
loan_data.drop(labels='id', axis=1, inplace=True)
loan_data.drop(labels='out_prncp_inv', axis=1, inplace=True)
loan_data.drop(labels='installment', axis=1, inplace=True)
loan_data.drop(labels='total_pymnt_inv', axis=1, inplace=True)
loan_data.drop(labels='total_rec_prncp', axis=1, inplace=True)
loan_data.drop(labels='collection_recovery_fee', axis=1, inplace=True)
loan_data.drop(labels='total_rev_hi_lim', axis=1, inplace=True)
loan_data.drop(labels='total_acc', axis=1, inplace=True)
loan_data.drop(labels='mths_since_last_delinq', axis=1, inplace=True)
loan_data.drop(labels='total_rec_int', axis=1, inplace=True)
```

```python
loan_data.drop(labels='emp_title', axis=1, inplace=True)
```

```python
loan_data.drop(columns = ['member_id', 'sub_grade', 'url', 'desc', 'title',
                          'zip_code', 'next_pymnt_d', 'recoveries',
                          'total_rec_late_fee'], inplace = True)
```

```python
loan_data_corr = loan_data.copy()
loan_data_corr
```

```python
from sklearn import preprocessing
label_encoding = preprocessing.LabelEncoder()
loan_data_corr[''] = label_encoding.fit_transform(loan_data_corr['class'].astype(str))
loan_data_corr.head()
```
